```python
title=0
experience=1
field=2
years=2
for i in jobs:
    a_class= i.findAll('a',{'class':"css-o171kl"})
    span_class=i.find('div',{'class':'css-1rhj4yg'}).findAll('span')
    str=i.find('div',{'class':'css-1rhj4yg'}).text
    keywords=str.split(' · ')
    print("Job Title: ",a_class[title].text) #job title
    print("Offering Company: ",i.find('div',{'class':'css-1k5ee52'}).find('a').text)
    print("Located in: ",i.find('span').text)
    try:
        print("Date Posted: ",i.find('div',{'class':'css-1jldrig'}).text)
    except:
        print("Date Posted: ",i.find('div',{'class':'css-eg55jf'}).text)
    print("Work Schedule(Full/Part-Time): ",i.find('span',{'class':'css-uc9rga eoyjyou0'}).text)
    try:
        print("Work Model: ",i.find('span',{'class':'css-uofntu eoyjyou0'}).text)
    except:
        print("not stated.")
    print("Experience: ",a_class[experience].text)
    try:
        print("Years Of Experience: ",span_class[years].text)
    except:
        print("Years Of Experience: not stated.")
    print("Field: ",a_class[field].text)
    print("Keywords: ",keywords[3::])
    print()
##Note: a certain job (OrbaSoft) has a different class=css-eg55jf for the date than the other jobs class=css-1jldrig
#to solve that issue i used a try-except statement to avoid the error
```

```python
file= "scraping.txt"
f = open(file,"w")
for i in jobs:
    a_class= i.findAll('a',{'class':"css-o171kl"})
    span_class=i.find('div',{'class':'css-1rhj4yg'}).findAll('span')
    str=i.find('div',{'class':'css-1rhj4yg'}).text
    keywords=str.split(' · ')
    f.write("Job Title: "+a_class[title].text+"\n") #job title
    f.write("Offering Company: "+i.find('div',{'class':'css-1k5ee52'}).find('a').text+"\n")
    f.write("Located in: "+i.find('span').text+"\n")
    try:
        f.write("Date Posted: "+i.find('div',{'class':'css-1jldrig'}).text+"\n")
    except:
        f.write("Date Posted: "+i.find('div',{'class':'css-eg55jf'}).text+"\n")
    f.write("Work Schedule(Full/Part-Time): "+i.find('span',{'class':'css-uc9rga eoyjyou0'}).text+"\n")
    try:
        f.write("Work Model: ",i.find('span'+{'class':'css-uofntu eoyjyou0'}).text+"\n")
    except:
        f.write("not stated."+"\n")
    f.write("Experience: "+a_class[experience].text+"\n")
    try:
        f.write("Years Of Experience: "+span_class[years].text+"\n")
    except:
        f.write("Years Of Experience: not stated."+"\n")
    f.write("Field: "+a_class[field].text+"\n")
    f.write("Keywords: "+",".join(keywords[3::])+"\n")
    f.write("\n---------------------------------------\n")
f.close()
```

```python
import pandas as pd
```

```python
data = []

for i in jobs:
    a_class = i.findAll('a', {'class': "css-o171kl"})
    span_class = i.find('div', {'class': 'css-1rhj4yg'}).findAll('span')
    str_text = i.find('div', {'class': 'css-1rhj4yg'}).text
    keywords = str_text.split(' · ')

    job = {
        "Job Title": a_class[title].text,
        "Company": i.find('div', {'class': 'css-1k5ee52'}).find('a').text,
        "Location": i.find('span').text,
        "Date Posted": i.find('div', {'class': 'css-1jldrig'}).text
                        if i.find('div', {'class': 'css-1jldrig'})
                        else i.find('div', {'class': 'css-eg55jf'}).text,
        "Work Schedule": i.find('span', {'class': 'css-uc9rga eoyjyou0'}).text,
        "Work Model": i.find('span', {'class': 'css-uofntu eoyjyou0'}).text
                        if i.find('span', {'class': 'css-uofntu eoyjyou0'})
                        else "not stated",
        "Experience": a_class[experience].text,
        "Years Of Experience": span_class[years].text if len(span_class) > years else "not stated",
        "Field": a_class[field].text,
        "Keywords": ",".join(keywords[3::])
    }
    data.append(job)

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("scraping.csv", index=False, encoding="utf-8")

# Save to JSON
df.to_json("scraping.json", orient="records", indent=4, force_ascii=False)
```
